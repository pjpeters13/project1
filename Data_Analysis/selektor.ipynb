```python
# import dependencies
import ipywidgets as widgets
from IPython.display import display
from IPython.core.getipython import get_ipython

# List of available notebooks
notebooks = [
    "10_most_common_criminal_offenses.ipynb",
    "ofns_data_analysis.ipynb",
    "trendiness.ipynb",
    "YearlyStats.ipynb",
    "jake-prophet.ipynb"
    "crime_per_year.ipynb
    "burrough_data.ipynb
]

# Defining a custom line magic command to start the notebook
def run_notebook(notebook_path):
    """Running Jupyter notebook with IPython magic command `%run`."""
    get_ipython().run_line_magic('run', notebook_path)

# Creating a dropdown menu for selecting notebooks
dropdown = widgets.Dropdown(options=notebooks, description='Notebooks:')

# Creating a button to launch the selected notebook
button = widgets.Button(description="Run it!!!")

# A function that is called when the user clicks a button
def on_button_clicked(b):
    notebook_path = dropdown.value
    run_notebook(notebook_path)

# Adding an event handler to the button
button.on_click(on_button_clicked)

# Displaying dropdown menus and buttons
display(dropdown, button)
```
